## 1.7 Practice Exercise🎯: Generate Random Data to Simulate Rolling a Fair Six-Sided Dice

**Step 1:** Let's practice what we've learned so far by generating random data to simulate rolling a fair six-sided dice. A fair six-sided dice has an equal probability of landing on any of its six faces, numbered from 1 to 6. We will generate random data to simulate rolling the dice 1000 times and print the first few values.

````python
 # Generate random data to simulate rolling a fair six-sided dice 1000 times
import random

# Number of dice rolls
num_rolls = 1000

# Generate random data from a fair dice
dice_data = [random.randint(1, 6) for _ in range(num_rolls)]

# Print the first few values of the generated data
print(dice_data[:10])
````

````python
# Data type of the generated data
print(type(dice_data))  # Output: <class 'list'>
````

**Step 2:** Let's see how many times each face of the dice appears in the generated data. We can use the `Counter` class from the `collections` module to count the occurrences of each face.

````python
# Checking the distribution of the generated data
from collections import Counter # Counter is a class from the collections module that helps count hashable objects

# Count the occurrences of each face of the dice
dice_counts = Counter(dice_data)
dice_counts
````

````python
# Print the counts of each face
print("Counts of each face of the dice:")
for face, count in sorted(dice_counts.items()): # sorted() sorts the items by face number from 1 to 6
    print(f"Face {face}: {count} times equal to {count/num_rolls:.2%} of total rolls")
````

You can change the number of rolls by modifying the `num_rolls` variable. For example, if you want to simulate rolling the dice 5000 times, you can set `num_rolls = 5000`.


 **Step 3:** Now, let's visualize the distribution of the dice rolls using a histogram. We will use the `matplotlib` and `seaborn` libraries to create the histogram.

#### Handling Missing Values: `LotFrontage`

In this section, we explore how to deal with missing values in the `LotFrontage` column, which represents the linear feet of street connected to the property.

---

##### 🔍 Understanding `LotFrontage`
`LotFrontage` indicates the length of the property facing the street (in feet). It is an important feature in real estate as it can affect accessibility, zoning, and value.

---

##### 📊 Initial Observations

- `LotFrontage` has **489 missing values** (~17%). We need to handle these missing values appropriately. Dropping these rows may not be the best option!
- The values range from **21 to 313 feet**. (Information from `data.describe()`)

---
Let's explore the `HouseStyle` column to understand the context of these missing values. `HouseStyle` indicates the architectural style of the house, which can provide insights into the type of property and its street access.



````python
# Unique values of HouseStyle
data['HouseStyle'].unique()  # Display the unique values of the HouseStyle column
````

##### ⚠️ Why Not Fill with Zero?
- Filling with `0` is **not appropriate** because:
  - All `HouseStyle` entries (e.g., `1Story`, `2Story`, `SFoyer`) represent **standalone houses** — not apartments or townhouses, which by definition have street access.
  - A `0` would incorrectly imply the house has **no street frontage**, which could mislead any predictive model.
  - It introduces artificial outliers or extreme values.

---

##### ✅ Recommended Approach: Fill with Median with `HouseStyle` Context
- We will fill missing `LotFrontage` values with the **median** of `LotFrontage` for each unique `HouseStyle`.
- This approach respects the architectural context and maintains the


````python
# Create a copy of the original LotFrontage, so we can compare before and after imputation
data['LotFrontage_imputed_basic'] = data['LotFrontage']

# Loop through each unique HouseStyle and fill missing LotFrontage with the median
for style in data['HouseStyle'].unique():
    median_value = data[data['HouseStyle'] == style]['LotFrontage'].median()
    mask = (data['HouseStyle'] == style) & (data['LotFrontage_imputed_basic'].isnull())
    data.loc[mask, 'LotFrontage_imputed_basic'] = median_value

# Count any remaining missing values
data['LotFrontage_imputed_basic'].isnull().sum()
````

````python
print(median_value)
````

````python
# Summarize the imputed medians
imputed_summary_basic = data.groupby('HouseStyle')['LotFrontage_imputed_basic'].median()

imputed_summary_basic
````

Notice that the imputed values are now more consistent with the distribution of `LotFrontage` for each `HouseStyle`. This method respects the architectural context and avoids introducing misleading values.

#### 3.2.3 The Relationship Between Continious Predictors

Multicollinearity is a phenomenon in regression analysis where two or more predictor variables are highly correlated with each other. It indicates a strong linear relationship between the predictors, which can cause issues in the regression model. When multicollinearity is present, it becomes challenging to distinguish the individual effects of the correlated predictors on the response variable.

To address multicollinearity, various techniques can be employed, such as removing one of the correlated predictors or incorporating an interaction effect. The correlation matrix below provides an overview of the correlations between each factor.

**Correlation** is a way to measure how two things are related to each other.

**Formula to Calculate the Correlation**
$$
r = \frac{\sum (x_i - \bar{x})(y_i - \bar{y})}
         {\sqrt{\sum (x_i - \bar{x})^2} \cdot \sqrt{\sum (y_i - \bar{y})^2}}
$$

What it means:
- $r$: Pearson correlation coefficient (ranges from -1 to +1)
- $x_i$: The $i^{th}$ value of variable X
- $y_i$: The $i^{th}$ value of variable Y
- $\bar{x}$: The mean (average) of all values in variable X
- $\bar{y}$: The mean (average) of all values in variable Y
- $\sum$: The summation symbol (sum of all values)
- $(x_i - \bar{x})$: How far each X value is from the mean of X
- $(y_i - \bar{y})$: How far each Y value is from the mean of Y
- $\sqrt{\sum (x_i - \bar{x})^2}$: Standard deviation of X (spread of X values)
- $\sqrt{\sum (y_i - \bar{y})^2}$: Standard deviation of Y (spread of Y values)


**In simple words:**

The formula compares how the values of X and Y move together.

If one thing goes up and the other tends to go up too, that’s a positive correlation (like height and weight — taller people tend to weigh more).

If one thing goes up and the other tends to go down, that’s a negative correlation (like how more time spent watching TV might mean lower test scores).

If there’s no clear pattern, then there’s no correlation.

Correlation is usually shown as a number between -1 and +1:

+1 means a perfect positive relationship.

-1 means a perfect negative relationship.

0 means no relationship at all.

---
***IMPORTANT: It doesn NOT mean one causes the other — it just shows they move together (or not).***

📊 **Introduction to ANOVA F-Statistic**

When working with datasets that include categorical variables (like GarageType, HouseStyle) and a continuous target variable (like SalePrice), it's important to evaluate whether those categorical features have a meaningful impact on the target.

One useful statistical method for this is the Analysis of Variance (ANOVA).

ANOVA, short for Analysis of Variance, tests whether the mean of a continuous variable (like SalePrice) is significantly different across two or more categories of a categorical variable.

In simple terms:

Does the target variable (e.g., SalePrice) vary between different groups of a feature (e.g., GarageType)?

📐 **How It Works?**

ANOVA calculates an F-statistic, which measures the ratio of:

Between-group variance (how much the group means differ from the overall mean)

Within-group variance (how much values differ inside each group)

$𝐹 = \frac{Variance Between Groups}{Variance Within Groups}$

A higher F-statistic suggests the group means are more different, indicating a stronger relationship between the categorical feature and the target variable.

If the F-statistic is close to 1, the group means are likely similar, and the feature has little predictive value.

✅ **When to Use It?**

You have categorical features and want to know if they affect a continuous target variable

You want a quick ranking of which categorical features are worth exploring for predictive modeling

````python
from scipy.stats import f_oneway
import pandas as pd

# Dictionary to store both F-statistic and p-value
anova_results = {}

for col in categorical_vars:
    # Group the log sale price by category
    groups = [group['LSalePrice'] for _, group in data.groupby(col)]

    # Only run ANOVA if we have at least 2 groups
    if len(groups) > 1:
        f_stat, p_val = f_oneway(*groups)
        anova_results[col] = {'F-statistic': f_stat, 'p-value': p_val}

# Convert to DataFrame and sort by F-statistic
anova_df = pd.DataFrame(anova_results).T.sort_values(by='F-statistic', ascending=False)

# Show result
print(anova_df.round(3))
````

**Why do we use `*groups` in `f_oneway(*groups)`?**

The * operator is called the "unpacking operator" in Python.

What’s happening: groups is a list of Series, like this:

`groups = [group1, group2, group3, group4...]`
The `f_oneway()` function does not accept a list — it expects each group as a separate argument, like:

`f_oneway(group1, group2, group3, group4)`

So we use `*groups` to unpack the list into separate arguments:

`f_oneway(*groups)`

This is equivalent to:

`f_oneway(groups[0], groups[1], groups[2], ...)`


````python
# Convert the Series to a 2D DataFrame (1 row, many columns)
plt.figure(figsize=(12, 5))
sns.heatmap(anova_df[['F-statistic']].T, annot=True, fmt=".1f", cmap="Blues") # Using .T to transpose the df
plt.title("ANOVA F-statistics: Categorical vs LSalePrice")
plt.yticks(rotation=0)  # Keep feature label readable
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels
plt.tight_layout()
plt.show()
````
